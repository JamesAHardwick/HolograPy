In this example script we give examples of how many of the functions available in the HolograPy

```python
import numpy as np
import plotly.graph_objects as go
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import HolograPy as hp
```

Firstly, we define the basic physical parameters of the sound fields we will be working with. Speed of sound, operating frequency and wavelength, etc...

```python
# physics params
c0 = 343.  # [m/s]
f = 40000
lam = c0/f
k = 2*np.pi/lam
```

Now we define a set of points for our transducers, modulator and evaluation plane usi

```python
# Defining a transducer plane
t_v1, t_v2 = [1,0,0], [0,1,0]
t_spacingx, t_spacingy = 10/1000, 10/1000
t_nx, t_ny = 16, 16
t_origin = [-t_spacingx * t_nx/2, -t_spacingy * t_ny/2, 0]
t_radius = 4.5/1000
t_normals = [0, 0, 1] # all transducers are facing in +z direction.

t_points = hp.compute_grid_2d(t_origin, t_v1, t_v2, t_spacingx, t_spacingy, t_nx, t_ny)

# Defining a modulation plane
m_v1, m_v2 = [1,0,0], [0,1,0]
m_spacingx, m_spacingy = lam/2, lam/2
m_nx, m_ny = 32, 32
m_origin = [-m_spacingx * m_nx/2, -m_spacingy * m_ny/2, 100/1000] # 100mm above (0, 0, 0)
m_normals = [0, 0, 1] # all modulator elements are facing in +z direction.
m_areas = m_spacingx*m_spacingy # all modulator elements have the same area in this example

m_points = hp.compute_grid_2d(m_origin, m_v1, m_v2, m_spacingx, m_spacingy, m_nx, m_ny)

# Defining an evaluation plane
e_v1, e_v2 = [1,0,0], [0,1,0]
e_spacingx, e_spacingy = 2/1000, 2/1000
e_nx, e_ny = 64, 64
e_origin = [-e_spacingx * e_nx/2, -e_spacingy * e_ny/2, 200/1000] # 200mm above (0, 0, 0)

e_points = hp.compute_grid_2d(e_origin, e_v1, e_v2, e_spacingx, e_spacingy, e_nx, e_ny)
```

To visualise these planes, we can use the plotly package. We have functions to visualise arrays of transducers, generic points or the edges of a plane.

```python
fig = go.Figure()

hp.plot_transducers_plotly(fig, t_points, t_radius, label="Transducers")
hp.plot_plane_points_plotly(fig, m_points, color='salmon', label="Modulation points")
hp.plot_plane_edges_plotly(fig, e_points, color='slateblue', label="Evalutaion_plane")

fig.update_layout(
    scene=dict(
        xaxis=dict(showgrid=False),
        yaxis=dict(showgrid=False),
        zaxis=dict(showgrid=False),
        # aspectmode='data'  # equal scaling for all axes
    ),
    width=700,
    height=700,
    showlegend=True
)

fig.show()
```

With our coordinate system defined, we can now find the forward model that maps wave behaviour as the waves travel from the transducers to the modulation plane. For this we use the PM_forward_model function.

```python
# Find the forward model mapping between the transducer coords and modulation plane coords.
H_tran_mod = hp.PM_forward_model(m_points, t_points, t_normals, k)

# We assume all transducers are activated with an amplitude of 1 and a phase of pi.
P_tran = np.ones((t_nx * t_ny, 1)) * np.exp(1j * np.pi)

print("H_tran_mod:", H_tran_mod.shape)
print("P_tran:", P_tran.shape)

# To find the pressure incoming to the modulation plane from the transducers, we do:
P_mod_incoming = H_tran_mod @ P_tran

print("P_mod_incoming:", P_mod_incoming.shape)
```

```python
# Define a target

targets = hp.char_to_array(['U'], 60, (e_nx, e_ny), font_file="Arial.ttf")

for i, ch in enumerate(targets):
    plt.subplot(1, len(targets), i+1)
    plt.imshow(ch, cmap='gray')
    plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import gaussian_filter

def wgs(A, b, K=500, incoming_field=None, smooth_sigma=None, alpha=0.8):
    """
    Weighted Gerchbergâ€“Saxton (WGS) solver optimized for letters or structured targets.

    Parameters
    ----------
    A : np.ndarray, shape (M, N), complex
        Forward model matrix mapping the hologram plane to the target plane.
    b : np.ndarray, shape (M, 1) or (M,), complex
        Target complex field (desired amplitude and phase in the target plane).
    K : int, optional
        Number of iterations (default 500).
    incoming_field : np.ndarray, shape (N, 1) or (N,), complex, optional
        Incoming complex field at the hologram plane. Defaults to uniform plane wave.
    smooth_sigma : float or None, optional
        Standard deviation for Gaussian smoothing of the target amplitude.
        If None or 0, no smoothing is applied. 1 = maximum smoothing (good for letters).
    alpha : float, optional
        Relaxation parameter for hologram update (0 < alpha <= 1).

    Returns
    -------
    x : np.ndarray, shape (N, 1), complex
        Final hologram field: same amplitude as incoming_field, optimized phase.
    y : np.ndarray, shape (M, 1), complex
        Forward propagated field in the target plane.
    """
    M, N = A.shape
    b = b.reshape((M, 1))

    # Incoming field
    if incoming_field is None:
        inc = np.ones((N, 1), dtype=np.complex128)
    else:
        inc = incoming_field.reshape((N, 1))

    AT = np.conj(A).T
    x = inc.copy()

    # Create mask for letter region (non-zero amplitude)
    mask = (np.abs(b) > 0).astype(np.float32)

    # Optionally smooth mask
    if smooth_sigma and smooth_sigma > 0:
        mask = gaussian_filter(mask, sigma=smooth_sigma)

    # Flattened target amplitude
    target_amplitude = mask / np.max(mask)

    for _ in range(K):
        # Forward propagation
        y = A @ x  # preserves amplitude of incoming field

        # Replace amplitude in target plane with flattened amplitude in letter region
        p = target_amplitude * (y / np.abs(y))

        # Backward propagation
        r = AT @ p

        # Update hologram with relaxation, preserve incoming amplitude
        x_phase = r / np.abs(r)
        x = inc * ((1 - alpha) * (x / np.abs(x)) + alpha * x_phase)
        x = x / np.max(np.abs(x)) * np.abs(inc)  # preserve amplitude scale of incoming field

    # Final forward propagation
    y = A @ x

    return x, y
```

```python
# Find the forward model mapping between the modulation plane coords and the evaluation plane coords.
H_mod_eval = hp.GF_forward_model(e_points, m_points, m_normals, m_areas, k)
```

```python
# Use WGS (Weighted Gerchberg-Saxton) to find the phase modulation required at the modulation plane:
complex_target = targets[0] * np.exp(1j * (targets[2] != 0).astype(int))
P_mod_outgoing, P_eval = wgs(H_mod_eval, complex_target, incoming_field=P_mod_incoming, smooth_sigma=1.0)

# Final field at hologram  plane
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3.5))
im = ax1.imshow(abs(P_mod_outgoing.reshape(m_nx, m_ny)))
cax = ax1.inset_axes([1.1, 0.1, 0.05, 0.8]) # [x0, y0, width, height]
plt.colorbar(im, cax=cax)

im=ax2.imshow(np.angle(P_mod_outgoing.reshape(m_nx, m_ny)))
cax = ax2.inset_axes([1.1, 0.1, 0.05, 0.8]) # [x0, y0, width, height]
plt.colorbar(im, cax=cax)
plt.show()

# Final propagated field at evaulation plane
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3.5))
im=ax1.imshow(abs(P_eval.reshape(e_nx, e_ny)))
cax = ax1.inset_axes([1.1, 0.1, 0.05, 0.8]) # [x0, y0, width, height]
plt.colorbar(im, cax=cax)

im=ax2.imshow(np.angle(P_eval.reshape(e_nx, e_ny)))
cax = ax2.inset_axes([1.1, 0.1, 0.05, 0.8]) # [x0, y0, width, height]
plt.colorbar(im, cax=cax)
plt.show()
```
